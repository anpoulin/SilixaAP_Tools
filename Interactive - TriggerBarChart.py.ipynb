```python
import matplotlib.pyplot as plt
import datetime

# Data
days_of_year = [292, 293, 294, 295, 296, 297, 298, 299, 300, 301, 302, 303, 304]
triggers = [202, 219, 147, 349, 314, 1243, 788, 1166, 1900, 1900, 603, 153, 529]

# Convert DOY to dates
year = 2024
dates = [datetime.datetime(year, 1, 1) + datetime.timedelta(days=doy - 1) for doy in days_of_year]
formatted_dates = [date.strftime('%Y/%m/%d') for date in dates]

# Create bar chart
plt.figure(figsize=(12, 6))
plt.bar(formatted_dates, triggers, color='skyblue')

# Add labels and title
plt.xlabel('Date (YYYY/MM/DD)')
plt.ylabel('Number of Triggers')
plt.title('Number of Triggers per Date')

# Rotate date labels for better readability
plt.xticks(rotation=45)

# Show plot
plt.show()
```
